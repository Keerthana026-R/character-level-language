```python
import re
import PyPDF2
import docx
from collections import Counter
```

```python
# Function to preprocess text (lowercase and remove non-alphabetic characters)
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    words = text.split()
    return words
```

```python
# Function to extract text from a PDF file
def extract_text_from_pdf(pdf_file_path):
    with open(pdf_file_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page_num in range(len(reader.pages)):
            text += reader.pages[page_num].extract_text()
    return text
```

```python
# Function to extract text from a DOCX file
def extract_text_from_docx(docx_file_path):
    doc = docx.Document(docx_file_path)
    text = ''
    for para in doc.paragraphs:
        text += para.text + '\n'
    return text
```

```python
# Function to extract text from a TXT file
def extract_text_from_txt(txt_file_path):
    with open(txt_file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    return text
```

```python
# Function to calculate unigrams (word frequencies)
def unigram_model(document):
    words = preprocess_text(document)
    unigram_counts = Counter(words)
    return unigram_counts
```

```python
# Function to handle different file types
def process_document(file_path):
    if file_path.endswith('.pdf'):
        text = extract_text_from_pdf(file_path)
    elif file_path.endswith('.docx'):
        text = extract_text_from_docx(file_path)
    elif file_path.endswith('.txt'):
        text = extract_text_from_txt(file_path)
    else:
        raise ValueError("Unsupported file format. Please use a PDF, DOCX, or TXT file.")
    return text
```

```python
# Main Program
file_path = '/content/Pirates Of The Caribbean - The Curse Of The Black Pearl-www.learnenglishteam.com.pdf'  # Replace with your document path
```

```python
# Extract the document text
document_text = process_document(file_path)
```

```python
# Generate unigram frequencies
unigrams = unigram_model(document_text)
```

```python
# Display the results
for word, freq in unigrams.items():
    print(f"'{word}': {freq}")
```
